```python

#Basic libraries
import numpy as np
import tensorflow as tf
print('Tensorflow version: ', tf.__version__)
import time

#Show images
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 20)        # size of images
plt.rcParams['image.interpolation'] = 'nearest'  # show exact image

# Select GPU
import os
os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"]="0"

data_path = '/tmp/cifar10/'
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

print('X train shape: ',x_train.shape)
print('y train shape: ',y_train.shape)
print('X test shape: ', x_test.shape)
print('y test shape: ', y_test.shape)

categories = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
# Examine the data
from os import listdir
from os.path import isfile, join
import random

fig = plt.figure()
n=0
for i, category in enumerate(categories):
  category_positions = np.where(y_test == i)[0]
  random.shuffle(category_positions)
  for position in category_positions[:5]: 
    n += 1
    a = fig.add_subplot(10,5,n)
    a.set_title(category)
    plt.imshow(x_test[position])
    

    
```

```python
# First model


#Inputs
images = tf.keras.layers.Input(batch_shape=(None, 32, 32, 3), dtype='float32', name='Images')

#First convolutional layer
conv11 = tf.keras.layers.Conv2D(32, (3,3), padding='same', activation='relu')(images)
conv12 = tf.keras.layers.Conv2D(32, (3,3), padding='same', activation='relu')(conv11)
pool1 = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))(conv12)
pool1d = tf.keras.layers.Dropout(0.25)(pool1)

#Second convolutional layer
conv21 = tf.keras.layers.Conv2D(32, (3,3), padding='same', activation='relu')(pool1d)
conv22 = tf.keras.layers.Conv2D(32, (3,3), padding='same', activation='relu')(conv21)
pool2 = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))(conv22)
pool2d = tf.keras.layers.Dropout(0.25)(pool2)

#Dense layers
flat = tf.keras.layers.Flatten(name='Flat_image')(pool2d)
dense1 = tf.keras.layers.Dense(512, activation='relu', name='Dense_1')(flat)
dense1d = tf.keras.layers.Dropout(0.5)(dense1)
output = tf.keras.layers.Dense(10, activation='softmax', name='Dense_output')(dense1d)


# Model Architecture defined
model1 = tf.keras.models.Model(inputs=images, outputs=output)
model1.summary()


# Select optimizer and compile model
opt = tf.keras.optimizers.RMSprop(lr=0.001, decay=1e-5)
model1.compile(loss='sparse_categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
```

```python
nb_train_samples = 50000
nb_test_samples = 10000
epochs = 35
batch_size = 128
```

```python
# subtract mean and normalize
#mean_image = np.mean(x_train, axis=0)
#x_train = np.array((x_train - mean_image) / 128, dtype=np.float32)
#x_test  = np.array((x_test  - mean_image) / 128, dtype=np.float32)
```

```python
# Augmentation for training
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1/255,
    shear_range=0.2,
    zoom_range=0.2,
    width_shift_range=0.1,  # randomly shift images horizontally (fraction of total width)
    height_shift_range=0.1,  # randomly shift images vertically (fraction of total height)
    horizontal_flip=True)

# Augmentation configuration we will use for testing:
# only rescaling
test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255)

```

```python
# test generator
g = test_datagen.flow(x_test, y_test, batch_size=1)

gen_data = next(g)
print(categories[gen_data[1][0][0]])

plt.imshow(gen_data[0][0])
```

```python
history1 = model1.fit_generator(
    train_datagen.flow(x_train,y_train, batch_size=batch_size),
    steps_per_epoch = nb_train_samples // batch_size,
    epochs = epochs,
    validation_data = test_datagen.flow(x_test, y_test, batch_size=batch_size),
    validation_steps = nb_test_samples // batch_size)
```
